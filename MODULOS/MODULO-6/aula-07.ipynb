```python
import pandas as pd

dados_baguncados = {
    "Meses": ["Jan", "Fev", "Fev", "Mar"], # 'Fev' duplicado!
    "Safra_KG": [1500, 2300, 2300, "1800"] # '1800' está como texto!
}

df_baguncado = pd.DataFrame(dados_baguncados)
print("--- BASE BAGUNÇADA ---")
print(df_baguncado)
```

```python
df_sem_duplicadas = df_baguncado.drop_duplicates()

print("--- BASE SEM DUPLICADAS ---")
print(df_sem_duplicadas)
```

```python
# Sobrescrevendo a coluna com a versão convertida
df_sem_duplicadas["Safra_KG"] = df_sem_duplicadas["Safra_KG"].astype(int)

# Tirando o Raio-X para confirmar se virou número
df_sem_duplicadas.info()
```
