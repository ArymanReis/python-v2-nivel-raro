```python
import pandas as pd

# Carregando a nossa base de dados
df = pd.read_csv("../DB/safra.csv")
```

```python
# Repare nos parênteses blindando cada condição!
condicao_and = (df['Safra_KG'] > 2000) & (df['Faturamento_RS'] > 25000)

super_safra = df[condicao_and]
super_safra.head()
```

```python
# Lembre-se: No Python, usamos dois iguais (==) para comparar se um texto é idêntico ao outro.
condicao_or = (df['Meses'] == 'Jan') | (df['Meses'] == 'Mar')

meses_especificos = df[condicao_or]
meses_especificos.head()
```
