```python
import pandas as pd

dados_brutos = {
    "cod_mes_ref": ["Jan", "Fev", "Mar", "Abr"],
    "vlr_lucro_liq": [20000, 7000, 45000, 32000]
}

df_final = pd.DataFrame(dados_brutos)
print("--- BASE COM NOMES RUINS ---")
print(df_final)
```

```python
# Passamos um dicionário: {"nome_velho": "nome_novo"}
df_final = df_final.rename(columns={
    "cod_mes_ref": "Meses", 
    "vlr_lucro_liq": "Lucro_Liquido"
})

print("--- BASE RENOMEADA ---")
print(df_final)
```

```python
# Ordenando pela coluna Lucro_Liquido. ascending=False significa "do maior para o menor"
df_ordenado = df_final.sort_values(by="Lucro_Liquido", ascending=False)

print("--- RANKING DE LUCRO ---")
print(df_ordenado)
```
