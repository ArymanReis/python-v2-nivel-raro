```python
import pandas as pd

dados_com_falha = {
    "Meses": ["Jan", "Fev", "Mar", "Abr"],
    "Safra_KG": [1500, 2300, None, 3100] # O sensor falhou aqui!
}

df_sujo = pd.DataFrame(dados_com_falha)
print("--- BASE SUJA ---")
print(df_sujo)
```

```python
df_limpo = df_sujo.dropna()

print("--- BASE APÓS DROPNA ---")
print(df_limpo)
```

```python
# O número dentro dos parênteses é o valor que vai tapar o buraco
df_preenchido = df_sujo.fillna(0)

print("--- BASE APÓS FILLNA ---")
print(df_preenchido)
```
