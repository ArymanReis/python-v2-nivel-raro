```python
import pandas as pd

dados_exportacao = {
    "Meses": ["Jan", "Fev", "Mar", "Abr"],
    "Tipo_Cafe": ["Arábica", "Robusta", "Arábica", "Arábica"],
    "Exportacao_Sacas": [1200, 800, 950, 1500]
}

df_cafe = pd.DataFrame(dados_exportacao)
print(df_cafe)
```

```python
primeira_linha = df_cafe.iloc[0]
print("--- PRIMEIRA LINHA EXATA ---")
print(primeira_linha)
```

```python
# Passo 1: Construímos a pergunta blindada com parênteses
filtro_arabica_elite = (df_cafe["Tipo_Cafe"] == "Arábica") & (df_cafe["Exportacao_Sacas"] > 1000)

# Passo 2: Usamos o .loc para aplicar a regra na tabela
resultado_final = df_cafe.loc[filtro_arabica_elite]

# Exibindo o ouro
print("--- CAFÉS ARÁBICA DE ALTA EXPORTAÇÃO ---")
print(resultado_final)
```
