```python
import pandas as pd

# Recebemos uma base apenas com receitas e despesas
dados_financeiros = {
    "Meses": ["Jan", "Fev", "Mar", "Abr"],
    "Faturamento_RS": [50000, 42000, 60000, 55000],
    "Custo_RS": [30000, 35000, 20000, 25000]
}

df_financas = pd.DataFrame(dados_financeiros)
print("--- BASE ORIGINAL ---")
print(df_financas)
```

```python
# Criando a nova coluna do zero
df_financas["Lucro_Liquido"] = df_financas["Faturamento_RS"] - df_financas["Custo_RS"]

print("--- BASE COM NOVA COLUNA (ENGENHARIA DE RECURSOS) ---")
print(df_financas)
```
