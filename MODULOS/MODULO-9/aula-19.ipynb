```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Cruzamento de dados da Fazenda
dados = {
    "Chuva_mm": [10, 25, 40, 60, 80],
    "Lucro_Mil": [15, 30, 45, 65, 85]
}
df_analise = pd.DataFrame(dados)

# 2. O Motor Estatístico (Seaborn)
sns.scatterplot(
    x="Chuva_mm",
    y="Lucro_Mil",
    data=df_analise,
    color="green",
    s=100  # Tamanho da bolinha
)

# 3. Exibição
plt.title("Correlação: Chuva vs Lucro")
plt.show()
```
