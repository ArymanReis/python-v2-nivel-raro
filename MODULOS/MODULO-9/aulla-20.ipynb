```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Base de dados
df_custos = pd.DataFrame({
    "Setor": ["TI", "Vendas", "RH", "Mkt"],
    "Custo": [25000, 12000, 8000, 15000]
})

# 2. Gráfico de Barras
sns.barplot(
    x="Setor",
    y="Custo",
    data=df_custos,
    color="steelblue"
)

# 3. Customização
plt.title("Custos por Departamento")
plt.xlabel("Setores")
plt.ylabel("Valor (R$)")

# 4. A foto em Alta Resolução
plt.savefig(
    "relatorio_custos.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python
# Simulando 10 vendas diferentes
df_vendas = pd.DataFrame({
    "Valor_Venda": [50, 55, 60, 50, 200, 65, 50, 45, 50, 300]
})

# Criando o Histograma
sns.histplot(
    data=df_vendas,
    x="Valor_Venda",
    bins=5,
    color="purple"
)

plt.title("Distribuição do Valor das Vendas")
plt.show()
```
