```python
import matplotlib.pyplot as plt

# 1. Os dados da fazenda (Eixos X e Y)
meses = ["Jan", "Fev", "Mar", "Abr"]
custos = [15000, 18000, 16000, 21000]

# 2. O traçado da linha
plt.plot(
    meses,
    custos,
    marker="o",
    color="red"
)

# 3. A tradução para o Diretor
plt.title("Evolução de Custos da Fazenda")
plt.xlabel("Meses do Ano")
plt.ylabel("Gasto em Reais (R$)")

# 4. Exibição do painel
plt.show()
```
