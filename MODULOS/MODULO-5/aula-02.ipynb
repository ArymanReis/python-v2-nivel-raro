```python
import numpy as np
```

```python
# Criando um Array do NumPy
safra_kg = np.array([1500, 2300, 1800, 3100, 2000])

print("Safra Original:", safra_kg)

# A Diretoria pediu para converter tudo de KG para Toneladas (dividir por 1000)
# Veja como não precisamos de um comando 'for'
safra_toneladas = safra_kg / 1000

print("Safra em Toneladas:", safra_toneladas)
```
