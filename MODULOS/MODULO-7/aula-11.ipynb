```python
precos = [1000, 2500, 3000, 1800, 4200]
precos_com_desconto = []

for p in precos:
    precos_com_desconto.append(p * 0.9) # Tirando 10%

print("Jeito Manual:", precos_com_desconto)
```

```python
precos = [1000, 2500, 3000, 1800, 4200]

# Sintaxe: [calculo for item in lista]
precos_com_desconto = [p * 0.9 for p in precos]

print("Jeito Alta Performance:", precos_com_desconto)
```

```python
# Sintaxe: [calculo for item in lista if condicao]
precos_elite = [p * 0.9 for p in precos if p > 2000]

print("Apenas Sacas Elite com Desconto:", precos_elite)
```
