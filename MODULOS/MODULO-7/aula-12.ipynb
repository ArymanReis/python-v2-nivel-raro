```python
import pandas as pd

# Simulando uma base que poderia ter 50 mil linhas
dados_exportacao = {"Saca_USD": [150, 200, 180, 220, 190]}

df_export = pd.DataFrame(dados_exportacao)
print("--- PREÇOS ORIGINAIS (USD) ---")
print(df_export)
```

```python
# O 'x' representa cada valor individual da coluna Saca_USD
df_export["Saca_BRL_Liq"] = df_export["Saca_USD"].apply(
    lambda x: (x * 5) * 0.9
)

print("--- PREÇOS CONVERTIDOS COM IMPOSTO (BRL) ---")
print(df_export)
```
