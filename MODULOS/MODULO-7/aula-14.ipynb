```python
import pandas as pd

dados = {
    "Data_Texto": ["2023-01-15", "2023-06-20", "2024-02-10"],
    "Sacas": [100, 150, 120]
}
df_tempo = pd.DataFrame(dados)

# Convertendo texto puro para o formato DateTime
df_tempo["Data_Real"] = pd.to_datetime(
    df_tempo["Data_Texto"]
)

print("--- TIPOS DE DADOS ---")
print(df_tempo.dtypes)
```

```python
# Usando o superpoder .dt para fatiar o tempo
df_tempo["Ano"] = df_tempo["Data_Real"].dt.year
df_tempo["Mes"] = df_tempo["Data_Real"].dt.month
df_tempo["Trimestre"] = (
    df_tempo["Data_Real"].dt.quarter
)

print("--- BASE COM SAZONALIDADE ---")
print(df_tempo)
```

```python
# Isolando apenas as safras de início de ano
filtro_q1 = df_tempo[df_tempo["Trimestre"] == 1]

print("--- APENAS 1º TRIMESTRE ---")
print(filtro_q1)
```
