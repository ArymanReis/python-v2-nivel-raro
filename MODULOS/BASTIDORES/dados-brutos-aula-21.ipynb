```python
import os
import pandas as pd

# 1. Sai da pasta DB (../) e cria a pasta dados_brutos junto com os Módulos
os.makedirs("../dados_brutos", exist_ok=True)

# 2. Criando 2 CSVs falsos de filiais
pd.DataFrame({"ID_Cliente": [1, 2], "Valor": [100, 250]}).to_csv("../dados_brutos/vendas_SP.csv", index=False)
pd.DataFrame({"ID_Cliente": [3, 1], "Valor": [300, 150]}).to_csv("../dados_brutos/vendas_RJ.csv", index=False)

# 3. Criando 1 JSON falso simulando uma API Web
pd.DataFrame({
    "ID_Cliente": [1, 2, 3],
    "Nome": ["Ana", "Bruno", "Carlos"],
    "Status": ["VIP", "Comum", "Comum"]
}).to_json("../dados_brutos/clientes.json", orient="records")

print("Fábrica concluída! Os arquivos foram gerados na pasta 'dados_brutos'.")
```
