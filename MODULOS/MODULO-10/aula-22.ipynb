```python
import pandas as pd

# (Simulando as tabelas que extraímos no Dia 21)
df_vendas = pd.DataFrame({"ID_Cliente": [1, 2, 3, 1], "Valor": [100, 250, 300, 150]})
df_clientes = pd.DataFrame({"ID_Cliente": [1, 2, 3], "Nome": ["Ana", "Bruno", "Carlos"], "Status": ["VIP", "Comum", "Comum"]})

print("Iniciando a Transformação (ETL - Fase 2)...")

# 1. Cruzando os dados (Merge)
df_consolidado = pd.merge(
    df_vendas,
    df_clientes,
    on="ID_Cliente",
    how="inner"
)

# 2. Respondendo à Diretoria (GroupBy)
# "Quanto cada categoria de cliente gastou?"
df_resumo = df_consolidado.groupby("Status")["Valor"].sum().reset_index()

print("\n--- Tabela Final Cruzada ---")
print(df_consolidado)

print("\n--- Faturamento por Status ---")
print(df_resumo)
```
