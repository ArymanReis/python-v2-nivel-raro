```python
import sqlite3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Iniciando a Carga (ETL - Fase 3)...")

# (Simulando o resumo gerado no Dia 22)
df_resumo = pd.DataFrame({"Status": ["Comum", "VIP"], "Valor": [450, 100]})

# -----------------------------------------
# 1. CARGA NO BANCO DE DADOS (Cofre Seguro)
# -----------------------------------------
conexao = sqlite3.connect("../DB/empresa.db")

df_resumo.to_sql(
    "relatorio_vendas",
    conexao,
    if_exists="replace",
    index=False
)
conexao.close()

# -----------------------------------------
# 2. GERAÇÃO DO DASHBOARD (A Entrega)
# -----------------------------------------
sns.barplot(
    x="Status",
    y="Valor",
    data=df_resumo,
    color="gold"
)

plt.title("Faturamento Consolidado por Cliente")
plt.xlabel("Categoria")
plt.ylabel("Total (R$)")

# Salvando a entrega final em PDF
plt.savefig(
    "dashboard_diretoria.pdf",
    dpi=300,
    bbox_inches="tight"
)
plt.show()

print("Pipeline ETL Concluído! Dados salvos e PDF gerado.")
```
