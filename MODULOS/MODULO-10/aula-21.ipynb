```python
import pandas as pd
import glob

print("Iniciando a Extração (ETL - Fase 1)...")

# -----------------------------------------
# 1. Extração do Sistema Antigo (Múltiplos CSVs)
# -----------------------------------------
caminho_csv = "../dados_brutos/vendas_*.csv"
lista_arquivos = glob.glob(caminho_csv)

tabelas_vendas = []
for arquivo in lista_arquivos:
    df_temp = pd.read_csv(arquivo)
    tabelas_vendas.append(df_temp)

# Empilhando as 50 filiais em uma tabela só
df_vendas_total = pd.concat(
    tabelas_vendas, 
    ignore_index=True
)

# -----------------------------------------
# 2. Extração do Sistema Web Moderno (JSON)
# -----------------------------------------
caminho_json = "../dados_brutos/clientes.json"
df_clientes = pd.read_json(caminho_json)

print(f"Sucesso! {len(df_vendas_total)} Vendas importadas.")
print(f"Sucesso! {len(df_clientes)} Clientes importados.")
```
