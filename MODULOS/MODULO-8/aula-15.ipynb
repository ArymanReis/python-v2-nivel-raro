```python
import sqlite3

# 1. Bate na porta (se o banco não existir, ele cria)
conexao = sqlite3.connect("../DB/empresa.db")

# 2. Chama o operário
cursor = conexao.cursor()

# 3. Executa um comando SQL (criar tabela)
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Clientes (
        ID INTEGER PRIMARY KEY,
        Nome TEXT,
        Setor TEXT
    )
''')
conexao.commit() # Salva a alteração
print("Banco conectado e tabela criada!")
```

```python
# Injetando uma linha na tabela usando SQL puro
cursor.execute('''
    INSERT INTO Clientes (Nome, Setor)
    VALUES ('TechCorp', 'Tecnologia')
''')
conexao.commit()
print("Dado inserido com sucesso no banco!")
```

```python
# Buscando a informação
cursor.execute("SELECT * FROM Clientes")
resultado = cursor.fetchall() # Traz tudo

print("--- RESULTADO DA BUSCA ---")
for linha in resultado:
    print(linha)

# 4. Encerramento: Tranca a porta do banco
conexao.close()
```
