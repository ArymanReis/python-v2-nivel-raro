```python
import pandas as pd
import sqlite3

# 1. Simulando nossa base limpa no Pandas
dados = {
    "Vendedor": ["Ana", "Bruno", "Carlos"],
    "Faturamento": [15000, 22000, 18500]
}
df_vendas = pd.DataFrame(dados)

# 2. Abrindo a porta do banco de dados
conexao = sqlite3.connect("empresa.db")

# 3. A Via Expressa: DataFrame para SQL
df_vendas.to_sql(
    name="Vendas_Q1",
    con=conexao,
    if_exists="replace",
    index=False
)

print("Carga de dados concluída com sucesso!")

# 4. Sempre tranque a porta
conexao.close()
```
