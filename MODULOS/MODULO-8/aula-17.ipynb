```python
import pandas as pd
import sqlite3

# =======================================================
# PASSO 0: PREPARANDO O TERRENO (CRIANDO O BANCO FALSO)
# =======================================================
conexao = sqlite3.connect("../DB/agronegocio.db")

# Fabricando a Tabela de Produção
df_prod = pd.DataFrame({
    "Data": ["2024-01-01", "2024-01-02", "2024-01-03"],
    "Sacas_Colhidas": [100, 120, 90]
})
df_prod.to_sql("Producao", conexao, if_exists="replace", index=False)

# Fabricando a Tabela de Clima
df_clima = pd.DataFrame({
    "Data": ["2024-01-01", "2024-01-02", "2024-01-03"],
    "Chuva_mm": [15.2, 0.0, 30.5]
})
df_clima.to_sql("Clima", conexao, if_exists="replace", index=False)


# =======================================================
# PASSO 1: A EXTRAÇÃO E O CRUZAMENTO (O SEU DIA 17)
# =======================================================
query_cruzamento = '''
    SELECT 
        P.Data, 
        P.Sacas_Colhidas, 
        C.Chuva_mm
    FROM Producao P
    INNER JOIN Clima C
        ON P.Data = C.Data
'''

# O Motor de Extração Pandas
df_analise = pd.read_sql_query(
    query_cruzamento, 
    conexao
)

print("--- BASE CRUZADA PARA ANÁLISE ---")
print(df_analise)

# Tranca a porta
conexao.close()
```
